**We will make countplot and piechart of ->"Segment","category","sub-category"**



```python
#no. of segmemt in data
ax=sns.countplot(x="Segment", data = df)
for bars in ax.containers:
    ax.bar_label(bars,label_type="edge")
```

```python
#pie chart for "segment"
segment_count = df["Segment"].value_counts()
#lets take the segment
segment = segment_count.index.tolist()
#lets take the count values of each segment
count = segment_count.values.tolist()
# Using Seaborn's pastel color palette
colors = sns.color_palette('pastel')[0:4]  

#making pie chart
plt.pie(count,labels = segment, autopct='%1.1f%%')
plt.show()
```

```python
#no. of products in category
ax = sns.countplot(x="Category", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#pie chart for"Category".
category_count = df["Category"].value_counts()
#let's take category count
category =  category_count.index.tolist()
#values of category count
count = category_count.values.tolist()

#making pie chart
plt.pie(count, labels = category, autopct = '%1.1f%%')
plt.show()
```

# Categorical- categorical

* heatmap, cluster map -> category, ship mode
* pairplot -> df

```python
# heatmap

# Create a cross-tabulation
cross_tab = pd.crosstab(df["Category"], df["Ship Mode"])

# Create the heatmap
sns.heatmap(cross_tab)

# Set labels and title
plt.xlabel("Ship Mode")
plt.ylabel("Category")
plt.title("Cross-tabulation of Category and Ship Mode")

# Show the plot
plt.show()
```

from the above we can see that most of the shipping is done on the same day

```python
# clustermap

# Create a cross-tabulation
cross_tab = pd.crosstab(df["Category"], df["Ship Mode"])

# Create the clustermap
sns.clustermap(cross_tab)

# Set labels and title
plt.xlabel("Ship Mode")
plt.ylabel("Category")
plt.title("Cross-tabulation of Category and Ship Mode")

# Show the plot
plt.show()
```

```python
#pair plot 
sns.pairplot(df, hue = "Category")
```

# Plot using aggregate function

```python

result1 = df.groupby('Category')['Postal Code'].agg(['mean', 'max', 'min']).reset_index()
result2 = df.groupby('Category')['Sales'].agg(['mean', 'max', 'min']).reset_index()

fig, axes = plt.subplots(2, 3, figsize=(40, 16))

sns.barplot(ax=axes[0, 0], data=result1, x='Category', y='min')
axes[0, 0].set_title('Min Postal Code of Different Category')
axes[0, 0].set_xlabel('Category')
axes[0, 0].set_ylabel('Min')

sns.barplot(ax=axes[0, 1], data=result1, x='Category', y='mean')
axes[0, 1].set_title('Mean Postal Code of Different Category')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Mean')

sns.barplot(ax=axes[0, 2], data=result1, x='Category', y='max')
axes[0, 2].set_title('Max Postal Code of Different Category')
axes[0, 2].set_xlabel('Category')
axes[0, 2].set_ylabel('Max')

sns.barplot(ax=axes[1, 0], data=result2, x='Category', y='min')
axes[1, 0].set_title('Min Sales of Different Category')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Min')

sns.barplot(ax=axes[1, 1], data=result2, x='Category', y='mean')
axes[1, 1].set_title('Mean Sales of Different Category')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Mean')

sns.barplot(ax=axes[1, 2], data=result2, x='Category', y='max')
axes[1, 2].set_title('Max Sales of Different Category')
axes[1, 2].set_xlabel('Category')
axes[1, 2].set_ylabel('Max')

plt.tight_layout()
plt.show()
```

# Plotting using groupby

```python
g1 = df.groupby(['Order Date'])['Sales'].sum().reset_index()
g1 = g1.sort_index(ascending=False)

g2 = df.groupby(['Order Date'])['Postal Code'].sum().reset_index()
g2 = g2.sort_index(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.lineplot(ax=axes[0], data=g1, x='Order Date', y='Sales')
axes[0].set_title('Total Sales in Order Date')

sns.lineplot(ax=axes[1], data=g2, x='Order Date', y='Postal Code')
axes[1].set_title('Total Postal Code in Order Date')

plt.tight_layout()
plt.show()
```
